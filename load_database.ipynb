```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('db/crypto-analysis.db')
c = conn.cursor()
```

```python
# load the data into a Pandas DataFrame
lunarCrush = pd.read_csv('./output/lunarCrush_data.csv')
coinbase = pd.read_csv('./output/cbpro_data.csv')

# write the data to a sqlite table
lunarCrush.to_sql('lunarCrush', conn, if_exists='replace', index = False)
coinbase.to_sql('coinbase', conn, if_exists='replace', index = False)
```

```python
pd.read_sql('''SELECT * FROM lunarCrush''', conn)
```

```python
pd.read_sql('''SELECT * FROM coinbase''', conn)
```

```python
pd.read_sql('''
    SELECT 
        coinbase.symbol, 
        coinbase.date, 
        coinbase.close, 
        lunarCrush.social_dominance
    FROM coinbase
    LEFT JOIN lunarCrush 
        ON lunarCrush.symbol = coinbase.symbol
        AND lunarCrush.time = coinbase.date
    ''',
    conn
)
```

```python
df = pd.read_sql("""
    SELECT 
        coinbase.symbol, 
        coinbase.date, 
        coinbase.close, 
        lunarCrush.social_dominance
    FROM coinbase
    LEFT JOIN lunarCrush 
        ON lunarCrush.symbol = coinbase.symbol
        AND lunarCrush.time = coinbase.date
    WHERE coinbase.symbol IN ('BTC','ETH') AND coinbase.date BETWEEN '2021-01-01' AND '2021-08-15'
    """,
    conn
)
```
